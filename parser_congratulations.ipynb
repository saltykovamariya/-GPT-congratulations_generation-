## Collection of congratulations for the subsequent training of the GPT 3 model

```python
from selenium.webdriver import Chrome 
from bs4 import BeautifulSoup
import requests
```

```python
# window for selenium
browser = Chrome("/Users/masha/chromedriver")
```

```python
url = "https://zedra.ru/po-imenam/"
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

## Get links

```python
links = []
```

```python
for link in soup.findAll('a'):
    if 's-dnem-rozhdeniya' in link.get('href'):
        links.append('https://zedra.ru/po-imenam' + link.get('href'))
```

```python
links = links[4:]
```

```python
links
```

## Get text

```python
all_text = ''
```

```python
for link in links:
    browser.get(link)
    poems=browser.find_elements_by_xpath\
    ('//div[@class="col-xs-12 col-sm-12 col-md-6 col-lg-6 shortstory_default"]')
    for i in range(len(poems)):
        all_text += browser.find_elements_by_xpath\
        ('//div[@class="col-xs-12 col-sm-12 col-md-6 col-lg-6 shortstory_default"]')[i].text
```

```python
all_text
```

```python
myText = open(r'/Users/masha/ds_bootcamp/learning/14-NLP/[GPT]congratulations_generation\congr_text.txt','w')
myText.write(all_text)
myText.close()
```
